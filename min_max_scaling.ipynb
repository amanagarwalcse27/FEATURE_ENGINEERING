```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\AMAN\OneDrive\Desktop\python ml\Datasets\wine_data.csv",header=None,usecols=[0,1,2])
df.columns=['Class label', 'Alcohol', 'Malic acid']
df.head()
```

```python
df['Class label'].value_counts()
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
colordict = {1:"Red",2:"Green",3:"Blue"}
sns.scatterplot(x= df['Alcohol'],y= df['Malic acid'],hue= df['Class label'],palette= colordict)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,1]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

```python
x1 = X_train['Alcohol']
y1 = X_train['Malic acid']

plt.subplot(1,2,1) #1 row, 2 column, 1 row
plt.scatter(x1,y1,marker= 'o')
plt.title("BEFORE SCALING")

x2 = X_train_scaled['Alcohol']
y2 = X_train_scaled['Malic acid']
plt.subplot(1,2,2)
plt.scatter(x2,y2,marker= 'o')
plt.title("AFTER SCALING")
```

```python
# -------- First subplot --------
plt.subplot(1, 2, 1)   # (rows, cols, index)
sns.scatterplot(
    data=X_train,
    x="Alcohol",
    y="Malic acid",
    hue=y_train,
)
plt.title("Before Scaling")

# -------- Second subplot --------
plt.subplot(1, 2, 2)
sns.scatterplot(
    data=X_train_scaled,
    x="Alcohol",
    y="Malic acid",
    hue=y_train,
)
plt.title("After Scaling")
```
